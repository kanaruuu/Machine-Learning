```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

results = model.train(
   data='data_custom.yaml',
   imgsz=640,
   epochs=15,
   batch=8,
   name='yolov8n_mini_custom',
    plots=True)
```
