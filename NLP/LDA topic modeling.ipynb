## 2. Data preprocessing

```python
import re
from nltk import word_tokenize
from nltk.corpus import stopwords
import string
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
def preprocess(text: str) -> str:
    pattern = re.compile(r'[^а-я]+')
    text = text.lower()
    text = pattern.sub(' ', text).strip()

    word_list = word_tokenize(text)
    stopwords_list = set(stopwords.words('russian'))

    word_list = [word for word in word_list if word not in stopwords_list]
    word_list = [word for word in word_list if len(word) > 2]
    word_list = [word for word in word_list if word not in string.punctuation]

    lemma = WordNetLemmatizer()
    word_list = [lemma.lemmatize(word) for word in word_list]
    text= ' '.join(word_list)

    return text
```

```python
data_train['text'] = data_train['text'].apply(lambda x: preprocess(x))
data_train.head()
```

```python
data_test['text'] = data_test['text'].apply(lambda x: preprocess(x))
data_test.head()
```

## Vectorization with CountVectorizer

```python
! pip install bigartm
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from artm import BatchVectorizer
import artm
```

```python
vectorizer = CountVectorizer(ngram_range=(1,3), max_features=20000)

X_train = vectorizer.fit_transform(data_train['text'])
X_test = vectorizer.transform(data_test['text'])
```

```python
X_train.shape
```

```python
n_wd_train = X_train.todense().T
n_wd_test = X_test.todense().T
```

```python
n_wd_train.shape, n_wd_test.shape
```

```python
vocabulary_train = vectorizer.get_feature_names_out()
vocabulary_train
```

```python
batch_vectorizer_train = BatchVectorizer(data_format='bow_n_wd', n_wd=n_wd_train, vocabulary=vocabulary_train, target_folder='batches_train')
```

```python
vocabulary_test = vectorizer.get_feature_names_out()
vocabulary_test
```

```python
batch_vectorizer_test = BatchVectorizer(data_format='bow_n_wd', n_wd=n_wd_test, vocabulary=vocabulary_test, target_folder='batches_test')
```

## 3. Fit model LDA with BigARTM, find the optimal number of topics

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(20, 15))

num_topics_values = [[10, 15], [25, 50]]

for i in range(2):
    for j in range(2):
        num_topics = num_topics_values[i][j]
        lda_model = artm.LDA(num_topics=num_topics, alpha=0.1, beta=0.1, num_document_passes=5, dictionary=batch_vectorizer_train.dictionary)
        lda_model.fit_offline(batch_vectorizer_train, num_collection_passes=10)
        perplexity_values = lda_model.perplexity_value
        axs[i, j].plot(perplexity_values, marker='o', label=f'Перплексия: {lda_model.perplexity_last_value}')
        axs[i, j].set_title(f'Количество тем: {num_topics}')
        axs[i, j].set_xlabel('Номер итерации')
        axs[i, j].set_ylabel('Перплексия')
        axs[i, j].legend()

plt.tight_layout()
plt.show()
```

## The optimal number of topics is 50

## 5. Fit the model common view

```python
model = artm.ARTM(num_topics=50)
```

```python
model.regularizers.add(artm.SmoothSparsePhiRegularizer(name='smooth_sparse_phi_regularizer', tau=-4))
```

```python
model.scores.add(artm.PerplexityScore(name='perplexity_score')) # перплексия
model.scores.add(artm.TopTokensScore(name='top_tokens_score')) # топовые темы
```

```python
model.scores
```

```python
model.num_document_passes = 10
```

```python
model.initialize(dictionary=batch_vectorizer_train.dictionary)
```

```python
model.fit_offline(batch_vectorizer_train, num_collection_passes=10)
```

```python
model.score_tracker['perplexity_score'].last_value
```

```python
model.score_tracker['perplexity_score'].value
```

```python
model.score_tracker['top_tokens_score'].last_tokens
```

```python
perplexity_values = model.score_tracker['perplexity_score'].value

plt.plot(perplexity_values, marker='o')
plt.title(f'Change perplexity in training')
plt.xlabel('Iterations')
plt.ylabel('Perplexity')
plt.legend()
plt.show()
```
