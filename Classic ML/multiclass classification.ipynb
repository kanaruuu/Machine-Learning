# Downsampling

```python
from sklearn.utils import resample

df_majority_hight = data[data.Diabetes_012==0.0]
df_majority_medium = data[data.Diabetes_012==1.0]
frames = [df_majority_hight, df_majority_medium]
df_majority = pd.concat(frames)
df_minority = data[data.Diabetes_012==2.0]

df_majority_downsampled = resample(df_majority,  
                                 replace=True,    # sample without replacement
                                 n_samples=35097,     # to match minority class
                                 random_state=123) # reproducible results

# Combine minority class with downsampled majority class
df_downsampled = pd.concat([df_majority_downsampled, df_minority])

df_down=df_downsampled.sample(frac=1)

X_down = df_down.drop("Diabetes_012", axis=1)
Y_down = df_down["Diabetes_012"]
```

# NB, Metrics
*Precision;
Recall;
F1-score;
Accuracy;
ROC AUC score;*

```python
def accuracy(test, test_prediction):
    correct = 0
    for i in range (len(test)):
        if test[i] == test_prediction[i]:
            correct += 1
    return (correct / len(test))
def Conf_matrix2(test, pred):
    zz=0
    zo=0
    zt=0
    oz=0
    oo=0
    ot=0
    tz=0
    to=0
    tt=0
    test = list(test)
    for i in range(len(test)):
        if pred[i]==0 and test[i]==0:
            zz+=1
        if pred[i]== 0 and test[i] == 1:
            zo+=1
        if pred[i] == 0 and test[i] == 2:
            zt+=1
            
        if pred[i]==1 and int(test[i])==0:
            oz+=1
        if pred[i]== 1 and test[i] == 1:
            oo+=1
        if pred[i] == 1 and test[i] == 2:
            ot+=1
            
        if pred[i]==2 and test[i]==0:
            tz+=1
        if pred[i]== 2 and test[i] == 1:
            to+=1
        if pred[i] == 2 and test[i] == 2:
            tt+=1
    matrix = {"0": [zz, zo, zt],
             "1":[oz, oo, ot],
             "2":[tz, to, tt]}
    return matrix
        
```

```python
def recall(y_test, y_pred):
    # РЕАЛЬНОСТЬ / ПРОГНОЗЫ
    m=dict(Conf_matrix2(y_test, y_pred))
    pzeros=m['0']
    pones=m['1']
    ptwos=m['2']
    
    zerop = (pzeros[0]) / (pzeros[0] + pzeros[1] + pzeros[2])
    onep = (pones[1]) / (pones[1] + pones[2] + pones[0])
    twop = (ptwos[2]) / (ptwos[2] + ptwos[0] + ptwos[1])
    
    return [[0, zerop],
            [1, onep],
            [2, twop]]

def precision(y_test, y_pred):
    # РЕАЛЬНОСТЬ / ПРОГНОЗЫ
    m=dict(Conf_matrix2(y_test, y_pred))
    pzeros=m['0']
    pones=m['1']
    ptwos=m['2']
    
    zerop = (pzeros[0]) / (pzeros[0] + pones[0] + ptwos[0])
    onep = (pones[1]) / (pzeros[1] + ptwos[1] + pones[1])
    twop = (ptwos[2]) / (pzeros[2] + pones[2] + ptwos[2])
    
    return [[0, zerop],
            [1, onep],
            [2, twop]]

def F1_score(test, test_prediction):
    B = 1
    prec = precision(test, test_prediction)
    rec = recall(test, test_prediction)
    return [[0, (1 + B ** 2) * prec[0][1] * rec[0][1] / ((B ** 2 * prec[0][1]) + rec[0][1])],
            [1, (1 + B ** 2) * prec[1][1] * rec[1][1] / ((B ** 2 * prec[1][1]) + rec[1][1])],
            [2, (1 + B ** 2) * prec[2][1] * rec[2][1] / ((B ** 2 * prec[2][1]) + rec[2][1])],]

```

# KNN

```python
import math

def euclidean_distance(data1, data2):
    distance = 0
    for i in range (len(data1) - 1):
        distance += (data1[i] - data2[i]) ** 2
    return math.sqrt(distance)

def get_neighbors(train, test, k=1):
    distances = [(train[i][-1], euclidean_distance(train[i], test))
                  for i in range (len(train))]
    distances.sort(key=lambda elem: elem[1])
    
    neighbors = [distances[i][0] for i in range (k)]
    return neighbors

def prediction(neighbors):
    count = {}
    for instance in neighbors:
        if instance in count:
            count[instance] +=1
        else :
            count[instance] = 1
    target = max(count.items(), key=lambda x: x[1])[0]
    return target

def accuracy2(test, test_prediction):
    correct = 0
    for i in range (len(test)):
        if test[i][-1] == test_prediction[i]:
            correct += 1
    return (correct / len(test))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3)
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()
X_train = X_train.to_numpy()
X_test = X_test.to_numpy()

y_train = y_train[0:2000]
y_test = y_test[0:2000]
X_train = X_train[0:2000, :]
X_test = X_test[0:2000, :]

x_train_concat = np.concatenate((X_train, y_train.reshape(-1, 1)), axis=1)
x_test_concat = np.concatenate((X_test, y_test.reshape(-1, 1)), axis=1)
print(f'X_train shape: {x_train_concat.shape},\n'
      f'X_test shape: {x_test_concat.shape}')

predictions = []
for x in range (len(x_test_concat)):
    neighbors = get_neighbors(x_train_concat, x_test_concat[x], k=5)
    result = prediction(neighbors)
    predictions.append(result)
#     print(f'predicted = {result}, actual = {x_test_concat[x][-1]}') # если есть интерес посмотреть, какие конкретно прогнозы некорректны

print(f'My accuracy: {accuracy2(x_test_concat, predictions)}')
```

# NB

```python
from scipy.stats import norm
class NaiveBayes:
    def __init__(self, X_train, X_test, y_train):
        self.X_train = X_train
        self.X_test = X_test
        self.y_train = y_train

    def fit(self):
      means = self.X_train.groupby(self.y_train).apply(np.mean)
      stds = self.X_train.groupby(self.y_train).apply(np.std)
      probs = self.X_train.groupby(self.y_train).apply(lambda x: len(x)) / self.X_train.shape[0]

      y_pred = []
      for elem in range(self.X_test.shape[0]):
          p = {}
          for cl in np.unique(self.y_train):
              p[cl] = probs.iloc[int(cl)]
              for index, param in enumerate(self.X_test.iloc[int(elem)]):
                  p[cl] *= norm.pdf(param, means.iloc[int(cl), int(index)], stds.iloc[int(cl), int(index)])
          y_pred.append(pd.Series(p).values.argmax())
      return y_pred

    
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3)

model = NaiveBayes(X_train, X_test, y_train)
y_pred = model.fit()

print(accuracy_score(y_test, y_pred))
```
