Разбиваем поля с датами запуска и дедлайна(они являются полями, контатенирующими в себе несколько значений) на год, месяц, день и приводим их к численному типу

```python
df['Launched_year'] = df['Launched'].str.split('-').str[0]
df['Launched_month'] = df['Launched'].str.split('-').str[1]
df['Launched_day'] = df['Launched'].str.split('-').str[2].str[0:2]

df['Deadline_year'] = df['Deadline'].str.split('-').str[0]
df['Deadline_month'] = df['Deadline'].str.split('-').str[1]
df['Deadline_day'] = df['Deadline'].str.split('-').str[2]

df['Launched_year'] = pd.to_numeric(df['Launched_year'])
df['Launched_month'] = pd.to_numeric(df['Launched_month'])
df['Launched_day'] = pd.to_numeric(df['Launched_day'])

df['Deadline_year'] = pd.to_numeric(df['Deadline_year'])
df['Deadline_month'] = pd.to_numeric(df['Deadline_month'])
df['Deadline_day'] = pd.to_numeric(df['Deadline_day'])

df
```

Удалим столбцы с датами дедлайна и запуска, оставшимися после предыдущего преобразования, а также название проекта, являющееся уникальным для каждого объекта

```python
df = df.drop(columns = ['Deadline', 'Launched', 'Name'])
```

Приведем признак состояния проекта к численному типу:  
0 - проект отменен  
1 - проект провалился  
2 - проект приостановлен  
3 - проект запущен  
4 - проект успешен

```python
df.loc[df['State'] == 'Canceled', 'State'] = '0'
df.loc[df['State'] == 'Failed', 'State'] = '1'
df.loc[df['State'] == 'Suspended', 'State'] = '2'
df.loc[df['State'] == 'Live', 'State'] = '3'
df.loc[df['State'] == 'Successful', 'State'] = '4'
df['State'] = pd.to_numeric(df['State'])
```

Взглянем на количество уникальных строк в столбцах признаков, оставшихся нечисловыми

```python
print(df['Category'].unique().shape)
print(df['Subcategory'].unique().shape)
print(df['Country'].unique().shape)
```

Очень часто наиболее информативными признаками являются признаки с высокой дисперсией, посмотрим на показатели отклонения и понизим признаковое пространство

```python
df2.describe()
```

```python
df_t = pd.DataFrame(df2, columns=['Goal', 'Pledged', 'Backers', 'State'])
df_t
```

```python
df_t.to_csv("preprocessed_data.csv", index=False)
```

Такое понижение размерности признакового пространства позволяет сэкономить ресурсы при обучении модели и в ряде случаев повышает его качество

Визуализация данных

```python
import seaborn as sns
sns.heatmap(df_t.corr(), annot = True)
```

```python
from matplotlib.pyplot import figure
fig, axs = plt.subplots(nrows= 1 , ncols= 3 )

axs[0].scatter(df_t["Backers"], df_t["State"])
axs[1].scatter(df_t["Pledged"], df_t["State"])
axs[2].scatter(df_t["Goal"], df_t["State"])
```

```python
nMax = 10000

res = df_t.groupby('State').apply(lambda x: x.sample(n=min(nMax, len(x))))

print(res['State'].value_counts())
```

Разбиение данных на тестовую и обучающую выборки, используя стратификацию, т.к. наши классы не сбалансированы

```python
from sklearn.model_selection import train_test_split
y = res["State"]
X = res.drop(["State"], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

BaggingClassifier

```python
from sklearn.ensemble import BaggingClassifier, GradientBoostingClassifier, StackingClassifier, RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
bag_model = BaggingClassifier(n_estimators = 101).fit(X_train, y_train)
bag_model.score(X_test, y_test)
```

Неплохие показатели, учитывая, что у нас целых 5 классов

```python
pred_test = bag_model.predict(X_test)
print('Precision, Recall, F\n', classification_report(y_test, pred_test))
```

GradientBoostingClassifier

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
grid_space = {
  'criterion': ['friedman_mse', 'squared_error'],
  'n_estimators': np.arange(1, 150, 10),
  'max_depth': np.arange(2, 15, 1)
}

grid = RandomizedSearchCV(GradientBoostingClassifier(),grid_space,cv=3, scoring='accuracy')
```

```python
model_grid = grid.fit(X_train[:1000], y_train[:1000])
```

```python
model_grid.best_params_
```

```python
pred_test = model_grid.predict(X_test)
print(accuracy_score(y_test, pred_test))
```

```python
print('Precision, Recall, F\n', classification_report(y_test, pred_test))
```

K Nearest Neighbours

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score

knn_optimal = GridSearchCV(KNeighborsClassifier(),
                  {'n_neighbors': np.arange(5, 7, 1),
                  'p': np.arange(1, 2, 1),
                  "metric" : ["euclidean", "minkowski"]
                  },
                  ).fit(X_train, y_train)



pred_train = knn_optimal.predict(X_test)

print(classification_report(y_test, pred_train))

print("Accuracy :", accuracy_score(y_test, pred_train))
```

Отчет: была понижена размерность признакового пространства путем отбора признаков с максимальной дисперсией. После обучения моделей с подобором гиперпараметров, лучшей с небольшим отрывом оказалась модель К ближайших соседей

```python
%%writefile app.py
import pickle

import pandas as pd
import numpy as np
import streamlit as st
from sklearn.metrics import confusion_matrix, classification_report


def main():

    page = st.sidebar.selectbox(
        "Выберите страницу",
        ["Описание задачи и данных", "Запрос к модели"]
    )

    if page == "Описание задачи и данных":
        st.title("Описание задачи и данных")
        st.write("Выберите страницу слева")

        st.header("Описание задачи")
        st.markdown("""
        Модель KNN обучена на датасете из 73k+ записей \n
        В разных ситуациях люди и компании могут заниматься краундфайдингом - искать финансирование своим проектам.\n
        Набор данных содержит информацию о различных проектах, которая включает в себя финансовую цель, количество людей, которые помогли проекту финансово,
        а также количество людей, обязавшихся оплатить некий взнос при успешном завершении краудфандинга.
        Необходимо предсказать одно из пяти состояний проекта в данный момент времени ('Canceled', 'Failed', 'Live', 'Successful', 'Suspended').""")

        st.header("Описание данных")
        st.markdown("""Предоставленные данные:\n
вещественные признаки:
* Goal - финансовая цель
* Pledged -  количество людей, обязавшиеся оплатить некий взнос при успешном завершении краудфандинга
* Backers - количество людей, которые помогли проекту финансово
* State - состояний проекта в данный момент времени:\n
'Canceled'\n
'Failed',\n
'Live',\n
'Successful',\n
'Suspended'\n
""")

    elif page == "Запрос к модели":
        model = load_model("knn.pkl")
        test_data = pd.read_csv("preprocessed_data.csv")
        #st.title(*test_data)
        y =  test_data["State"]
        X = test_data.drop(["State"], axis=1)

        st.title("Запрос к модели")
        st.write("Выберите страницу слева")
        request = st.selectbox(
            "Выберите запрос",
            ["Сделать прогноз", "Метрики", "Первые 20 предсказанных значений"]
        )

        if request == "Метрики":
            st.header("Метрики")
            y_pred = model.predict(X)
            cr=classification_report(y, y_pred)
            st.write(cr)
            #'Classification Report: ',cr
            st.write(confusion_matrix(y, y_pred))
        elif request == "Первые 20 предсказанных значений":
            st.header("Первые 20 предсказанных значений")
            y_pred = model.predict(X.iloc[:20,:])
            for item in y_pred:
                st.write(f"{item:.2f}")
        elif request == "Сделать прогноз":
            st.header("Сделать прогноз")

            Goal = st.number_input("Goal", 0, 70000000000)
            Pledge = st.number_input("Pledge", 0, 300000000)
            Backers = st.number_input("Backers", 0, 30000000)

            if st.button('Предсказать'):
                data = [Goal, Pledge, Backers]
                data = np.array(data).reshape((1, -1))
                pred = model.predict(data)

                if pred[0]==0:
                    st.write("Canceled")

                elif pred[0]==1:
                    st.write("Failed")

                elif pred[0]==2:
                    st.write("Suspended")

                elif pred[0]==3:
                    st.write("Live")

                elif pred[0]==4:
                    st.write("Successful")


            else:
                pass



@st.cache_data
def load_model(path_to_file):
    with open(path_to_file, 'rb') as model_file:
        model = pickle.load(model_file)
    return model


@st.cache_data
def load_test_data(path_to_file):
    df = pd.read_csv(path_to_file, sep=";")
    return df


if __name__ == "__main__":
    main()
```
